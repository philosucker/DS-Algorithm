# 문10) 셸 정렬이란? 구현방법
단순삽입정렬을 개선한 정렬방법.  
단순삽입정렬은 배열의 모든 원소를 뒤에서 앞으로 차례대로 계속 비교해나간다.  

    단점 : 중반이후까지 어느정도 정렬이 된 상태라면, 후반부의 작은 원소를 초반부에 삽입해야 하는 경우 시간이 매우 오래 걸린다.   
            개선되지 않은 단순교환, 단순선택 정렬과 비슷하게 걸릴 것이다. O(n^2)
    
    개선방법 : 비교작업을 "배열의 바로 이웃하는 원소"단위로 하지 않고, 
             배열에서 똑같은 길이(h=h*3+1)만큼 떨어진 원소들을 뽑아, 
                 앞에서부터 뒤로 한쌍씩 비교해가며 단순삽입정렬을 하고, shift해서 반복하고

                 앞에서부터 뒤로 한쌍씩 비교해가며 단순삽입정렬을 하고, shift 해서 반복하고
             마지막엔 배열에서 1만큼 떨어진 원소들을 뽑아 
                 i.e. 맨앞에서부터 한쌍씩, 오른쪽으로 한칸씩 이동해가면서 비교해가며 단순삽입정렬을 하면
             
             정렬횟수는 늘어나지만, 원소의 이동횟수는 줄어들어 효율적이다.  
                 셸 정렬 시간복잡도 : O(n^1.25)

```python
def insertion_sort(a: MutableSequence) -> None:
    """단순 삽입 정렬"""
    n = len(a)
    for i in range(1, n):
        j = i       #   j는 배열의 두번째 인덱스부터 마지막 인덱스를 순차적으로 참조
        tmp = a[i]  # tmp는 배열의 두번째 원소부터 마지막 원소를 순차적으로 참조
        while j > 0 and a[j - 1] > tmp:  # j가 0보다 크고, a[i-1]이 a[i]보다 크면 
            a[j] = a[j - 1]                           # a[i] 는 a[i-1] 참조
            j -= 1                                    # j = j-1
        a[j] = tmp                                    # a[i-1] = a[i] 참조 
                   # 즉 i값을 계속 바꿔주기 위해 i.e. 오름차순 삽입을 마치고 다시 for문을 돌면서 뒤 원소를 앞에 정렬된 부분에 오름차순 삽입을 해주기 위해  
                                #  j가 i를 참조하게 하고
                                #  tmp 변수가 a[i]를 참조하게 해서 삽입시 사용한다.
    return a
```

```python
from typing import MutableSequence


def shell_sort(a: MutableSequence) -> None:
    """셸 정렬"""
    n = len(a)
    h = n // 2
    while h > 0:
        for i in range(h, n):            # i는 h부터 마지막 원소까지 순차적으로 참조
            j = i - h                    # j는 뒤로 건너뛴 h만큼 거꾸로 앞으로 간 i-h를 참조
            tmp = a[i]                   # tmp는 a[i]를 참조 
                                             # (앞쪽 원소가 뒤쪽 원소보다 큰 경우, 앞쪽 원소를 참조하는 변수가 뒤쪽 원소값 tmp를 참조하게 해서 바꿀 용도)
            while j >= 0 and a[j] > tmp: # j가 0보다 크거나 같고(맨 앞 원소도 비교해야하므로 그냥 단순삽입정렬에서 j>0와 달리 j>=0, 
                                           # 앞쪽 원소가 뒤쪽 원소보다 크면
                a[j + h] = a[j]           #  뒤쪽 원소를 참조하는 a[j+h] (i.e. a[i]) 는 앞쪽 원소를 참조하는 a[j] 값 참조
                j -= h                   # 앞쪽 원소를 참조하는 j는 현재 위치에서 h만큼 다시 앞으로 간 인덱스값을 참조, 
                 # while문에 다시 들어갔을 때, 만약 그 인덱스 값이 0보다 크거나 같다면, 
                 # 그 인덱스의 원소값이(i.e. 현재 위치에서 h만큼 앞에 있는 원소가) 뒤쪽원소 tmp보다 큰지 검사, 이런 식으로 h만큼 계속 거슬러 올라가서 다 검사
                 # 검사해서 앞쪽이 크면 뒤쪽이 앞쪽 참조하게 다 바꾸고
                 
            a[j + h] = tmp               # h만큼 거슬러 올라가 맨 앞까지 가면, tmp를 참조해 작은 값을 앞으로 삽입
        h //= 2
```

```python
from __future__ import annotations

def shell_sort(a: MutableSequence) -> None:
    """셸 정렬"""
    n = len(a)
    h = n // 2
    while h > 0:
        print(f'h: {h}')
        print(f'교환 전 a = {a}'); print()
        for i in range(h, n):
            j = i - h
            print(f'    교환 전 j = i - h : {j}'); print()
            tmp = a[i]
            print(f'    tmp: {tmp}'); print()
            while j >= 0 and a[j] > tmp:
                a[j + h] = a[j]
                j -= h
                
                print(f'               a[j + h] = a[j] 교환 후 a = {a}')
                print(f'               교환 후 j -= h : {j}'); print()
            a[j + h] = tmp
            print(f'    a[j + h] = tmp: a[{j} + {h}] = {tmp}')
            print(f'    교환 후 a = {a}'); print()
        h //= 2
    return a

shell_sort([8,1,4,2,7,6,3,5])
```

# 문11) 셸 정렬 개선방법, 구현

h값이 서로 배수관계일 경우,  
이전 h값에서 루프를 돌 때는 정렬이 안 일어나다가  
다음 h값에서 루프를 돌 때 정렬이 일어나는 경우가 생길 수 있다.  
  
**1. h가 배수관계가 생기지 않도록 해주면 된다.**  
h = h*3+1  
그러면 원소가 충분히 뒤섞여서 효율이 좋아진다.
  
**2. h의 초기값이 너무 커도 효과가 없다. 특정 길이 이상일 때부터 h 수열이 적용되게 해준다**  
배열 길이가 어느 정도 있을 때 셸정렬을 쓰면 효율이 좋다.

```python
def shell_sort1(a: MutableSequence) -> None:
    """셸 정렬"""
    n = len(a)
    h = n // 2
    while h > 0:
        for i in range(h, n):
            j = i - h                    
            tmp = a[i]
            while j >= 0 and a[j] > tmp:
                a[j + h] = a[j]          
                j -= h                  
            a[j + h] = tmp              
        h //= 2
```

```python
def shell_sort2(a: MutableSequence) -> None:
    """셸 정렬(h * 3 + 1의 수열 사용)"""
    n = len(a)
    h = 1

    while h < n // 9:     
        '''배열 길이가 18보다 작으면 h = 1,  
                     18 이상이면 h=4로 시작, 
                     45 이상이면 h=13로 시작, 
                     126 이상이면 h=40로 시작,
                     369 이상이면 h=121로 시작,
        '''
        h = h * 3 + 1

    while h > 0:
        for i in range(h, n):
            j = i - h
            tmp = a[i]
            while j >= 0 and a[j] > tmp:
                a[j + h] = a[j]
                j -= h
            a[j + h] = tmp
        h //= 3
```

```python
def shell_sort2(a: MutableSequence) -> None:
    """셸 정렬(h * 3 + 1의 수열 사용)"""
    n = len(a)
    h = 1

    while h < n // 9:
        h = h * 3 + 1

    while h > 0:
        print(f'h: {h}')
        print(f'교환 전 a = {a}'); print()
        for i in range(h, n):
            j = i - h
            print(f'    교환 전 j = i - h : {j}'); print()
            tmp = a[i]
            print(f'    tmp: {tmp}'); print()
            while j >= 0 and a[j] > tmp:
                a[j + h] = a[j]
                j -= h
                
                print(f'               a[j + h] = a[j] 교환 후 a = {a}')
                print(f'               교환 후 j -= h : {j}'); print()
            a[j + h] = tmp
            print(f'    a[j + h] = tmp: a[{j} + {h}] = {tmp}')
            print(f'    교환 후 a = {a}'); print()
        h //= 3
```
