# 문2) 팩토리얼 재귀함수 구현, 반복문으로도 구현 (예외처리)


```python
def factorial(n: int) -> int:
    if n < 0:
        raise ValueError("음수는 팩토리얼을 계산할 수 없습니다.")
    elif n == 0:
        return 1
    else:
        return n * factorial(n - 1)
    
n = int(input('출력할 팩토리얼 값을 입력하세요: '))
print(f'{n}의 팩토리얼은 {factorial(n)}입니다.')
```

```python
def factorial(n: int) -> int:
    try:
        if n > 0:
            return n * factorial(n - 1)
        elif n == 0:
            return 1
        else:
            raise ValueError
    except ValueError:
        print("음수는 입력할 수 없습니다")
        
n = int(input('출력할 팩토리얼 값을 입력하세요: '))
print(f'{n}의 팩토리얼은 {factorial(n)}입니다.')
```

```python
def factorial_iterative(n):
    if n < 0:
        raise ValueError("음수는 입력할 수 없습니다")
    else:
        answer = 1
        for i in range(1, n+1):
            answer = answer * i
        
    return answer

factorial_iterative(-1)
```

# 문5) GCD 구하는 함수 구현

```python
def gcd(x: int, y: int) -> int:
    """정숫값 x와 y의 최대 공약수를 반환
    x > y"""
    
    print(f'gcd(x:{x}, y:{y})')
    print()
    if y == 0: # 4. 작은수를 0으로 나눌 때까지 가면 (나눠 떨어지면) 
        print(f'x = {x}')
        return x # 5. 앞단계의 작은수가 두 수의 GCD
    
    else:
        print(f'x: {x},  y: {y},  x%y: {x%y}')  # 1. 큰수를 작은수로 나눔. 나눠 떨어지면 작은수가 두 수의 GCD
        print(f'gcd({y}, {x%y})') # 2. 안 나눠 떨어지면
        print()
        return gcd(y, x % y) # 3. 나눠 떨어질때까지 작은수를 나머지로 나눔. 


print('두 정숫값의 최대 공약수를 구합니다.')
print()
x = int(input('첫 번째 정숫값을 입력하세요.: '))
y = int(input('두 번째 정숫값을 입력하세요.: '))
print()

print(f'두 정숫값의 최대 공약수는 {gcd(x, y)}입니다.')
```

# 문8) 재귀 알고리즘을 비재귀 알고리즘으로 바꾸기 

#### 상향식 분석을 해보면 재귀를 반복문으로 바꿀 힌트를 얻을 수 있다

### 1. 꼬리 재귀 제거하기

```python
def recur(n: int) -> int:
    if n > 0:
        recur(n - 1)
        print(n)
        recur(n - 2)
```

위 recur 함수는 순수한 재귀다. recur 함수 안에 recur가 두개 있으므로  
  
여기서 마지막 재귀 recur(n-2)가 하는 일은  
n을 n-2로 바꾸고 함수의 시작지점(자신을 호출한 recur(n))으로 돌아가는 것이다.  
왜냐면 recur(n-2)가 코드의 마지막이므로  
따라서 마지막 재귀를 다음과 같이 바꿀 수 있다

```python
def recur(n: int) -> int:
    """꼬리 재귀를 제거한 함수 recur"""
    while n > 0:
        recur(n - 1)
        print(n)
        n = n - 2
```

### 2. 재귀 제거하기

꼬리재귀를 제거한 함수에서 남은 재귀는 recur(n-1) 이다.  
문제는 이 코드가 마지막이 아니라 뒤에 코드가 남아 있다는 것이다.  
다시 말해 recur(n-1)을 없애고 반복문으로 바꾸려면,  
먼저 n을 어딘가에 저장해 놨다가  
반복문이 돌고 난 후  
그 뒤 print(n) 과 n = n - 2가 실행되게 해야 한다.  
**스택을 사용하면 이 문제를 해결할 수 있다**

```python
def recur(n: int) -> int:
    """재귀를 제거한 recur() 함수"""
    
    s = list() # 스택을 사용
    
    while True:
        if n > 0:
            s.append(n)  # n을 스택에 저장해 두고
            n = n - 1    # recur(n-1) 을 반복문으로 돌린다. 그러면 스택에 계속 n-1된 n이 쌓인다. n이 4였다면, 4, 3, 2, 1 순으로 쌓인다.
            continue
        print(s)
        if len(s) > 0:   # 이제 스택에서 저장된 n들을, 재귀가 가장 깊이 들어갔을 때의 n부터 거꾸로 하나씩 뺀다. I.e. 1, 2, 3, 4 순으로
            n = s.pop()  # 그러려면 스택에서 pop을 해주면 된다.
            print(n)     # 원래 재귀 순서대로 n을 출력하고 (맨 처음 1을 출력)
            n = n - 2    # recur(n-2)가 하는 것처럼 n을 n-2로 바꾸고, 반복문의 처음으로 돌아간다.(함수의 처음으로 돌아간다)
            continue     # 1-2는 -1이므로 바로 두번째 if문으로 건너뛰어서 2를 pop 하고 2출력, 2-2도 0이므로 다시 두번째 if문으로 와서 3을 pop 하고 3출력
        break            # 3-2는 1이므로 첫번째 if문에 걸려서 스택에 1 추가. 그러면 4, 1이 된다. n은 1-1=0이 돼서 바로 두번째 if문으로 건너 뛰고
                         # 1을 pop하고 1출력, 1-2는 -1 이므로 두번째 if문으로 건너 뛰어서 마지막 원소 4를 pop하고 4출력, n=2가 돼서
                         # 다시 스택에 2, 1이 쌓이고, 그대로 두번째 if문으로 계속 건너 뛰어서 남은 1, 2 출력
x = int(input('정수값을 입력하세요.: '))

recur(x)
```

# 문9) 하노이의 탑이란? 하노이의 탑 구현

문제: 

     3개의 기둥을 이용해서

        (3) 중간기둥에 있는 원반그룹을 목표기둥으로 옮기면 된다.
        
        1과 3을 재귀함수로 만들면 된다.
        
            원반개수 > 1 이라는 조건을 걸어 재귀호출을 하면
                가장 작은 원반1과 두번째로 작은 원반2만 남은 그룹일 경우
                    1번 재귀에 들어가서 바로 (2)만 출력하고 나오고 1>2
                                                           1>3
                    2번 재귀에 들어가서 바로 (2)만 출력하고 나오고 2>3
                    위로 올라간다.
            
            no = 3 이면
            함수진입
            1번재귀 move(2, 1, 2)
                  1번재귀 move(1, 1, 3) 
                        원반1을 시작기둥에서 목표기둥으로 옮기고    V
                        나오고
                  원반2를 시작기둥에서 중간기둥으로 옮기고          V
                  2번재귀 move(1, 3, 2)
                         원반 1을 목표기둥에서 중간기둥으로 옮기고  V
                         나오고
                  나오고
            원반3을 시작기둥에서 목표기둥으로 옮기고                V
            2번재귀 move(2, 2, 3)
                  1번재귀 move(1, 2, 1)
                         원반 1을 중간기둥에서 시작기둥으로 옮기고  V
                         나오고
                  원반2를 중간기둥에서 목표기둥으로 옮기고          V
                  2번재귀 move(1, 1, 3) 
                         원반1을 시작기둥에서 목표기둥으로 옮기고   V
                         나오고
                  나오고
            함수종료  

```python
def move(no: int, x: int, y: int) -> None:
    """원반을 no개를 x 기둥에서 y 기둥으로 옮김"""

    if no > 1: 
        move(no - 1, x, 6 - x - y)
        
    print(f'원반 [{no}]을(를) {x}기둥에서 {y}기둥으로 옮깁니다.')
    
    if no > 1: 
        move(no - 1, 6 - x - y, y)

move(4, 1, 3)
```

```python
def put() -> None: 
    '''pos를 출력'''
    for i in range(8):
        print(f'{pos[i]:2}', end='')   # :2 앞에 공백 하나를 두고 두번째 칸에 pos[i]를 출력 
    print()
```

```python
def set(i: int) -> None:
    '''i열에 퀸을 배치'''
    for j in range(8):
        pos[i] = j  # i열의 j행에 퀸을 배치
        if i == 7 : # 맨 끝열부터 첫번째 행에서 마지막 행까지 퀸을 다 배치하고 나면
            put()   # 체스판 출력
        else:           
            set(i+1)   # 맨 끝열에서 첫번째 열로 빠져나오면서 체스판에 모든 위치에 퀸을 배치
            
set(6)
'''set(6)은
첫번째 for문에서 pos[6] = 0 을 하고
else로 빠져서 set(7)을 호출한다.
그러면 pos[7] = 0 부터 pos[7]=7 까지 for문이 돌고 빠져나와
첫번째 for문의 pos[6] = 1 을 하고
다시 else로 빠져서 set(7)을 호출,
반복하고 빠져나와
첫번째 for문의 pos[6] = 2 을 하고 이하 반복

그러면 set(0)부터 시작하면
pos[0] = 0을 하고
set(1) 을 호출
다시 pos[1] = 0을 하고
set(2) 를  호출
다시 pos[2] = 0을 하고... 반복
set(6) ... set(7)까지 간다
'''
```

#### 아이디어2 구현, 분기branching한정bounding법

각 행에 퀸을 1개만 배치한다

```python
pos = [0] * 8
flag = [False] * 8

def put() -> None:
    
    for i in range(8):
        print(f'{pos[i]:2}', end='')
    print()

def set(i: int) -> None:
    
    for j in range(8):
        if not flag[j]: #  i열 j행에 방문처리가 안 돼 있다면
            pos[i] = j  #  i열 j행에 퀸을 배치하고 
            if i == 7:
                put()
            else:
                flag[j] = True  #  i열 j행에 방문처리를 한 후 
                set(i+1)        # i+1열로 이동
                flag[j] = False # set(i+1)을 빠져나오면 i열 j행 위치에 방문처리 취소, 그래야 이전 열에서 방문처리 안 된 행을 비울 수 있음
set(6)

'''set(0)
 0 1 2 3 4 5 6 7
 0 1 2 3 4 5 7 6
 0 1 2 3 4 6 5 7
 0 1 2 3 4 6 7 5
 0 1 2 3 4 7 5 6
 0 1 2 3 4 7 6 5
 0 1 2 3 5 4 6 7
 0 1 2 3 5 4 7 6
 0 1 2 3 5 6 4 7
 0 1 2 3 5 6 7 4
 0 1 2 3 5 7 4 6
...
 7 6 5 4 2 1 3 0
 7 6 5 4 2 3 0 1
 7 6 5 4 2 3 1 0
 7 6 5 4 3 0 1 2
 7 6 5 4 3 0 2 1
 7 6 5 4 3 1 0 2
 7 6 5 4 3 1 2 0
 7 6 5 4 3 2 0 1
 7 6 5 4 3 2 1 0
 '''
```

### 아이디어3 구현
각 대각선에 퀸을 1개씩만 배치하도록 한정bounding 작업 추가 적용 

```python
pos = [0]*8            # 체스판에 퀸이 배치된 상태
flag_a = [False]*8     # 각 행에 퀸을 배치했는지 체크
flag_b = [False]*15    # 오른쪽 대각선 방향으로 퀸을 배치했는지 체크
flag_c = [False]*15    # 왼쪽 대각선 방향으로 퀸을 배치했는지 체크

def put() -> None:
    
    for i in range(8):
        print(f'{pos[i]:2}', end='')
    print()

def set(i: int) -> None:
    
    for j in range(8):
        if(not flag_a[j]
           and not flag_a[j]
           and not flag_b[i+j]      # /대각선 방향 i,j조합의 계산값이 같은 경우, 대각선 수 총 15개, i,j의 조합의 합은 0부터 14까지 가능
           and not flag_c[i-j+7]):  # \대각선 방향 i,j조합의 계산값이 같은 경우 (+7은 i-j가 음수일 수 있으므로 음수보정용)
            pos[i] = j
            if i == 7:
                put()
            else:
                flag_a[j] = flag_b[i+j] = flag_c[i-j+7] = True
                print(f'{i}--가로--{flag_a}');print(f'{i}/대각선/{flag_b}');print(f'{i}\\대각선\{flag_c}')
                set(i+1)
                flag_a[j] = flag_b[i+j] = flag_c[i-j+7] = False
                print("--------------------------------------------------------")
                print(f'{i}--가로--{flag_a}');print(f'{i}/대각선/{flag_b}');print(f'{i}\\대각선\{flag_c}')
                print("========================================================")
    
set(0)
```

```python
pos = [0] * 8          # 각 열에 배치한 퀸의 위치
flag_a = [False] * 8   # 각 행에 퀸을 배치했는지 체크
flag_b = [False] * 15  # 대각선 방향(↙↗)으로 퀸을 배치했는지 체크
flag_c = [False] * 15  # 대각선 방향( ↘↖)으로 퀸을 배치했는지 체크

def put() -> None:
    """퀸을 놓는 상황을 □와 ■로 출력"""
    for j in range(8):
        for i in range(8):
            print('■' if pos[i] == j else '□', end='')
        print()
    print()

def set(i: int) -> None:
    """i 열의 알맞은 위치에 퀸을 놓기"""
    for j in range(8):
        if(     not flag_a[j]           # j 행에 아직 퀸을 놓지 않았으면
            and not flag_b[i + j]       # 대각선 방향(↙↗)으로 퀸이 배치 되지 않았다면
            and not flag_c[i - j + 7]): # 대각선 방향( ↘↖)으로 퀸이 배치 되지 않았다면
            pos[i] = j          # 퀸을 j 행에 놓기
            if i == 7:          # 모든 열에 퀸을 배치하는 것을 완료
                put()
            else:
                flag_a[j] = flag_b[i + j] = flag_c[i - j + 7] = True
                set(i + 1)      # 다음 열에 퀸을 놓기
                flag_a[j] = flag_b[i + j] = flag_c[i - j + 7] = False

set(0)          # 0 열에 퀸을 놓기
```
