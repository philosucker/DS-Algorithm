파이썬의 collections.deque는 (원형)이중연결리스트로 구현되어 있다. (아마 커서연결리스트로. 프리리스트를 사용하는)

    덱 = 이중연결리스트
    원형 덱 = 원형이중연결리스트 (자료구조 책 8장 원형이중연결리스트 참고)

원형 큐 = 링버퍼 (링버퍼 자료구조 책 4장 큐 )

    원형으로 구현하는 이유는 뒤쪽으로 요소를 채우다가, 
    공간이 다 차게 되면 tail과 head를 연결해 앞쪽의 빈 공간을 활용하려는 의도인데
    포인터로 구현한 연결리스트에는 애초에 빈공간이라는 개념이 존재하지 않기 때문에 
    원형은 아무런 의미가 없다.
        
        프리리스트를 쓰면?
    
    원형의 이점을 살리기 위해서라면 연결리스트가 아닌 배열로 풀어야 한다.
        
        커서 연결리스트?
        
여기서 원형 이중 연결리스트로 원형 덱을 구현하는 건   
단순히 문제풀이를 위해 구현하는 것이다.  
즉 반쪽짜리 구현

```python
from collections import deque 
class MyCircularDeque:

    def __init__(self, k: int):
        self.deque = deque([], maxlen=k)
        self.size = k

    def insertFront(self, value: int) -> bool:
        if len(self.deque) < self.size:
            self.deque.appendleft(value)
            return True
        else:
            False

    def insertLast(self, value: int) -> bool:
        if len(self.deque) < self.size:
            self.deque.append(value)
            return True
        else:
            False        

    def deleteFront(self) -> bool:
        if self.deque:
            self.deque.popleft()
            return True
        else:
            False

    def deleteLast(self) -> bool:
        if self.deque:
            self.deque.pop()
            return True
        else: 
            False  

    def getFront(self) -> int:
        if self.deque:
            return self.deque[0]
        else:
            return -1
    
    def getRear(self) -> int:
        if self.deque:
            return self.deque[-1]
        else:
            return -1     

    def isEmpty(self) -> bool:
        if not self.deque:
            return True
        else:
            False

    def isFull(self) -> bool:
        if len(self.deque) == self.size:
            return True
        else:
            False       
        
```

```python
class MyCircularDeque:
    def __init__(self, k: int):
        """
        크기 k로 덱을 초기화합니다.
        """
        self.k = k
        self.deque = [0] * k
        self.front = 0
        self.rear = 0
        self.count = 0

    def insertFront(self, value: int) -> bool:
        """
        덱의 앞쪽에 아이템을 추가합니다. 성공적으로 추가하면 True를, 실패하면 False를 반환합니다.
        """
        if self.isFull():
            return False
        self.front = (self.front - 1) % self.k
        self.deque[self.front] = value
        self.count += 1
        return True

    def insertLast(self, value: int) -> bool:
        """
        덱의 뒤쪽에 아이템을 추가합니다. 성공적으로 추가하면 True를, 실패하면 False를 반환합니다.
        """
        if self.isFull():
            return False
        self.deque[self.rear] = value
        self.rear = (self.rear + 1) % self.k
        self.count += 1
        return True

    def deleteFront(self) -> bool:
        """
        덱의 앞쪽에서 아이템을 삭제합니다. 성공적으로 삭제하면 True를, 실패하면 False를 반환합니다.
        """
        if self.isEmpty():
            return False
        self.front = (self.front + 1) % self.k
        self.count -= 1
        return True

    def deleteLast(self) -> bool:
        """
        덱의 뒤쪽에서 아이템을 삭제합니다. 성공적으로 삭제하면 True를, 실패하면 False를 반환합니다.
        """
        if self.isEmpty():
            return False
        self.rear = (self.rear - 1) % self.k
        self.count -= 1
        return True

    def getFront(self) -> int:
        """
        덱의 앞쪽 아이템을 반환합니다. 덱이 비어있으면 -1을 반환합니다.
        """
        if self.isEmpty():
            return -1
        return self.deque[self.front]

    def getRear(self) -> int:
        """
        덱의 뒤쪽 아이템을 반환합니다. 덱이 비어있으면 -1을 반환합니다.
        """
        if self.isEmpty():
            return -1
        return self.deque[(self.rear - 1) % self.k]

    def isEmpty(self) -> bool:
        """
        덱이 비어있는지 확인합니다.
        """
        return self.count == 0

    def isFull(self) -> bool:
        """
        덱이 가득 찼는지 확인합니다.
        """
        return self.count == self.k
```

```python
class Node:
    def __init__(self, value=None):
        self.value = value
        self.next = None
        self.prev = None

class MyCircularDeque:
    def __init__(self, k: int):
        """
        크기 k로 덱을 초기화합니다.
        """
        self.head = None
        self.tail = None
        self.count = 0
        self.capacity = k

    def insertFront(self, value: int) -> bool:
        """
        덱의 앞쪽에 아이템을 추가합니다. 성공적으로 추가하면 True를, 실패하면 False를 반환합니다.
        """
        if self.isFull():
            return False
        new_node = Node(value)
        if self.isEmpty():
            self.head = self.tail = new_node
            self.head.next = self.head.prev = self.head
        else:
            new_node.next = self.head
            new_node.prev = self.tail
            self.head.prev = new_node
            self.tail.next = new_node
            self.head = new_node
        self.count += 1
        return True

    def insertLast(self, value: int) -> bool:
        """
        덱의 뒤쪽에 아이템을 추가합니다. 성공적으로 추가하면 True를, 실패하면 False를 반환합니다.
        """
        if self.isFull():
            return False
        new_node = Node(value)
        if self.isEmpty():
            self.head = self.tail = new_node
            self.head.next = self.head.prev = self.head
        else:
            new_node.next = self.head
            new_node.prev = self.tail
            self.tail.next = new_node
            self.head.prev = new_node
            self.tail = new_node
        self.count += 1
        return True

    def deleteFront(self) -> bool:
        """
        덱의 앞쪽에서 아이템을 삭제합니다. 성공적으로 삭제하면 True를, 실패하면 False를 반환합니다.
        """
        if self.isEmpty():
            return False
        if self.head == self.tail:
            self.head = self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = self.tail
            self.tail.next = self.head
        self.count -= 1
        return True

    def deleteLast(self) -> bool:
        """
        덱의 뒤쪽에서 아이템을 삭제합니다. 성공적으로 삭제하면 True를, 실패하면 False를 반환합니다.
        """
        if self.isEmpty():
            return False
        if self.head == self.tail:
            self.head = self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = self.head
            self.head.prev = self.tail
        self.count -= 1
        return True

    def getFront(self) -> int:
        """
        덱의 앞쪽 아이템을 반환합니다. 덱이 비어있으면 -1을 반환합니다.
        """
        if self.isEmpty():
            return -1
        return self.head.value

    def getRear(self) -> int:
        """
        덱의 뒤쪽 아이템을 반환합니다. 덱이 비어있으면 -1을 반환합니다.
        """
        if self.isEmpty():
            return -1
        return self.tail.value

    def isEmpty(self) -> bool:
        """
        덱이 비어있는지 확인합니다.
        """
        return self.count == 0

    def isFull(self) -> bool:
        """
        덱이 가득 찼는지 확인합니다.
        """
        return self.count == self.capacity
```

```python
class ListNode:
    def __init__(self, value=None):
        self.value = value
        self.right = None  # 다음 노드를 가리키는 포인터
        self.left = None   # 이전 노드를 가리키는 포인터

class MyCircularDeque:
    
    def __init__(self, k: int) -> None:

        self.head, self.tail = ListNode(None), ListNode(None)
        self.k, self.len = k, 0   # self.k : 최대길이  self.len : 현재 길이 
        self.head.right, self.tail.left = self.tail, self.head  # head.right 가 tail
                                                                # tail.left가 head

    def _add(self, node: ListNode, new: ListNode) -> None:
        
        n = node.right    # n = node.right
        node.right = new  # node.right = new  해도 위의 n = node.right는 그대로 
                                            # >>> 파알 8장 연결리스트 노트북 assignment응용 참고
        new.left, new.right = node, n
        n.left = new
        
    def insertLast(self, value: int) -> bool:
        
        if self.len == self.k:
            return False
        self.len += 1
        self._add(self.tail.left, ListNode(value))
        return True        
        
    def insertFront(self, value: int) -> bool:
        
        if self.len == self.k:
            return False
        self.len += 1
        self._add(self.head, ListNode(value))  # 빈 리스트였다면 head와 tail 더미노드 사이에 새 노드 삽입
        return True
    
    def _del(self, node: ListNode) -> None:
        
        n = node.right.right
        node.right = n   # delete_front의 경우         head의 right를 head 다음 다음 노드로
        n.left = node    #                            head 다음 다음 노드의 left를 헤드로 
        
        # n = node.right.right  # delete_last의 경우   
        # node.right = n                             tail의 이전 이전 노드의 right를 tail로 
        # n.left = node                              tail의 left를 tail의 이전 이전 노드로 
        
    def deleteFront(self) -> bool:
        
        if self.len == 0:
            return False
        self.len -= 1
        self._del(self.head)
        return True
    
    def deleteLast(self) -> bool:
        
        if self.len == 0:
            return False
        self.len -= 1
        self._del(self.tail.left.left)
        return True
    
    def getFront(self) -> int:
        
        return self.head.right.value if self.len else -1
    
    def getRear(self) -> int:
        
        return self.tail.left.value if self.len else -1
    
    def isEmpty(self) -> bool:
        
        return self.len == 0
    
    def isFull(self) -> bool:
        
        return self.len == self.k
        
```
