# 문3) 연결리스트의 특징

이처럼 자기자신과 같은 형의 인스턴스를 참조하는 필드가 있는 구조를 **자기참조형**이라고 한다.  
  
리스트를 배열로 구현하면 원소를 중간에 삽입하거나 삭제할 때 그 이후의 모든 원소가 다 이동해야한다

# search() 함수

인수로 주어진 data 와 값이 같은 노드를 검색해,  
해당 노드의 순서번호를 검색하는 함수 (헤드 노드의 번호는 0)  
  
검색 알고리즘은 선형 검색을 사용

```python
class LinkedList:
    
    
    def __init__(self) -> None: # 노드가 하나도 없는 빈 연결리스트를 생성
        
        self.no = 0             # 연결리스트에 등록되어 있는 노드의 개수
        self.head = None        # self.head: 헤드 노드를 참조하기 위한 필드 
                                    # head는 머리 노드에 대한 참조일 뿐, 노드 그 자체가 아니다 !!!!!!!!!!!!!!!!!!
        self.current = None     # self.current : 현재 주목하고 있는 노드를 참조하는 필드 
                                # 주목 포인터라고 부르자.
                                    # 검색, 삽입, 삭제, 이동한 뒤의 해당 노드를 주목 (책 338p 표 8-1)
        
    def __len__(self) -> int:
        
        return self.no
    
    def search(self, data: Any) -> int:

        cnt = 0
        ptr = self.head              # 머리 노드부터 검색, 별도의 포인터 변수 ptr을 사용하는게 핵심.
        while ptr is not None:       # 연결리스트 끝까지 검색 (꼬리노드의 next는 None이므로)
            if ptr.data == data:     # 입력된 data 와 노드의 data가 일치하면
                self.current = ptr   # self.current를 해당 노드로 맞추고 
                return cnt           # 해당 노드의 순서번호를 리턴 (맨 앞 노드 순서번호는 0)
            cnt+=1                      # 일치하지 않으면 순서번호 +1
            ptr = ptr.next              # 머리노드의 뒤쪽노드를 참조
        
        return -1
    
    def __contains__(self, data: Any) -> Bool:  # 이 매직메서드를 구현함으로써 연결리스트에 in 연산자를 적용할 수 있다
        
        return self.search(data) >= 0
```

# add_first() 함수  
연결리스트 맨 앞에 노드를 삽입하는 함수

```python
class LinkedList:
    
    
    def __init__(self) -> None: # 노드가 하나도 없는 빈 연결리스트를 생성
        
        self.no = 0             # 연결리스트에 등록되어 있는 노드의 개수
        self.head = None        # self.head: 헤드 노드를 참조하기 위한 필드 
                                    # head는 머리 노드에 대한 참조일 뿐, 노드 그 자체가 아니다 !!!!!!!!!!!!!!!!!!
        self.current = None     # self.current : 현재 주목하고 있는 노드를 참조하는 필드 
                                # 주목 포인터라고 부르자.
                                    # 검색, 삽입, 삭제, 이동한 뒤의 해당 노드를 주목 (책 338p 표 8-1)
        
    def __len__(self) -> int:
        
        return self.no
    
    def search(self, data: Any) -> int:

        cnt = 0
        ptr = self.head              # 머리 노드부터 검색, 별도의 포인터 변수 ptr을 사용하는게 핵심.
        while ptr is not None:       # 연결리스트 끝까지 검색 (꼬리노드의 next는 None이므로)
            if ptr.data == data:     # 입력된 data 와 노드의 data가 일치하면
                self.current = ptr   # self.current를 해당 노드로 맞추고 
                return cnt           # 해당 노드의 순서번호를 리턴 (맨 앞 노드 순서번호는 0)
            cnt+=1                      # 일치하지 않으면 순서번호 +1
            ptr = ptr.next              # 머리노드의 뒤쪽노드를 참조
        
        return -1
    
    def __contains__(self, data: Any) -> Bool:  # 이 매직메서드를 구현함으로써 연결리스트에 in 연산자를 적용할 수 있다
        
        return self.search(data) >= 0
    
    def add_first(self, data: Any) -> None:
        
        ptr = self.head                            # 이것도 헤드노드를 참조하는 별도의 포인터 변수 ptr을 사용하는 게 핵심
        self.head = self.current = Node(data, ptr) # 입력받은 data 와 기존 헤드를 다음 노드로 하는 새 노드 생성 
                                                   # 새로 생성된 노드를 self.head와 self.current가 참조하게 한다.
        self.no += 1    # 전체 노드 개수 업데이트                
```

```python
class LinkedList:
    
    
    def __init__(self) -> None: # 노드가 하나도 없는 빈 연결리스트를 생성
        
        self.no = 0             # 연결리스트에 등록되어 있는 노드의 개수
        self.head = None        # self.head: 헤드 노드를 참조하기 위한 필드 
                                    # head는 머리 노드에 대한 참조일 뿐, 노드 그 자체가 아니다 !!!!!!!!!!!!!!!!!!
        self.current = None     # self.current : 현재 주목하고 있는 노드를 참조하는 필드 
                                # 주목 포인터라고 부르자.
                                    # 검색, 삽입, 삭제, 이동한 뒤의 해당 노드를 주목 (책 338p 표 8-1)
        
    def __len__(self) -> int:
        
        return self.no
    
    def search(self, data: Any) -> int:

        cnt = 0
        ptr = self.head              # 머리 노드부터 검색, 별도의 포인터 변수 ptr을 사용하는게 핵심.
        while ptr is not None:       # 연결리스트 끝까지 검색 (꼬리노드의 next는 None이므로)
            if ptr.data == data:     # 입력된 data 와 노드의 data가 일치하면
                self.current = ptr   # self.current를 해당 노드로 맞추고 
                return cnt           # 해당 노드의 순서번호를 리턴 (맨 앞 노드 순서번호는 0)
            cnt+=1                      # 일치하지 않으면 순서번호 +1
            ptr = ptr.next              # 머리노드의 뒤쪽노드를 참조
        
        return -1
    
    def __contains__(self, data: Any) -> Bool:  # 이 매직메서드를 구현함으로써 연결리스트에 in 연산자를 적용할 수 있다
        
        return self.search(data) >= 0
    
    def add_first(self, data: Any) -> None:
        
        ptr = self.head                            # 이것도 헤드노드를 참조하는 별도의 포인터 변수 ptr을 사용하는 게 핵심
        self.head = self.current = Node(data, ptr) # 입력받은 data 와 기존 헤드를 다음 노드로 하는 새 노드 생성 
                                                   # 새로 생성된 노드를 self.head와 self.current가 참조하게 한다.
        self.no += 1    # 전체 노드 개수 업데이트                   
        
    def add_last(self, data: Any) -> None:
        
        if self.head is None:     # 연결리스트가 비어 있으면 add_first(data) 실행 
            self.add_first(data)  # !!!!!!!!!!!!!!!!!!!!클래스의 메서드 안에서 클래스 내부의 또 다른 메서드를 호출하는 법
            
        else:
            ptr = self.head              # 이것도 헤드노드를 참조하는 별도의 포인터 변수 ptr을 사용하는게 핵심
            while ptr.next is not None:  # 꼬리노드까지 간 다음 (i.e. ptr.next = None이 될때까지 간 다음)
                ptr = ptr.next
            ptr.next = self.current = Node(data, None)  # ptr은 꼬리노드를 참조 따라서 기존 ptr.next는 None이었을 텐데, 
                                                        # 이제 None이 아닌 새로 추가되는 노드를 참조하게 만든다
            self.no += 1   # 전체 노드 개수 업데이트      
```

```python
class LinkedList:
    
    
    def __init__(self) -> None: # 노드가 하나도 없는 빈 연결리스트를 생성
        
        self.no = 0             # 연결리스트에 등록되어 있는 노드의 개수
        self.head = None        # self.head: 헤드 노드를 참조하기 위한 필드 
                                    # head는 머리 노드에 대한 참조일 뿐, 노드 그 자체가 아니다 !!!!!!!!!!!!!!!!!!
        self.current = None     # self.current : 현재 주목하고 있는 노드를 참조하는 필드 
                                # 주목 포인터라고 부르자.
                                    # 검색, 삽입, 삭제, 이동한 뒤의 해당 노드를 주목 (책 338p 표 8-1)
        
    def __len__(self) -> int:
        
        return self.no
    
    def search(self, data: Any) -> int:

        cnt = 0
        ptr = self.head              # 머리 노드부터 검색, 별도의 포인터 변수 ptr을 사용하는게 핵심.
        while ptr is not None:       # 연결리스트 끝까지 검색 (꼬리노드의 next는 None이므로)
            if ptr.data == data:     # 입력된 data 와 노드의 data가 일치하면
                self.current = ptr   # self.current를 해당 노드로 맞추고 
                return cnt           # 해당 노드의 순서번호를 리턴 (맨 앞 노드 순서번호는 0)
            cnt+=1                      # 일치하지 않으면 순서번호 +1
            ptr = ptr.next              # 머리노드의 뒤쪽노드를 참조
        
        return -1
    
    def __contains__(self, data: Any) -> Bool:  # 이 매직메서드를 구현함으로써 연결리스트에 in 연산자를 적용할 수 있다
        
        return self.search(data) >= 0
    
    def add_first(self, data: Any) -> None:
        
        ptr = self.head                            # 이것도 헤드노드를 참조하는 별도의 포인터 변수 ptr을 사용하는 게 핵심
        self.head = self.current = Node(data, ptr) # 입력받은 data 와 기존 헤드를 다음 노드로 하는 새 노드 생성 
                                                   # 새로 생성된 노드를 self.head와 self.current가 참조하게 한다.
        self.no += 1    # 전체 노드 개수 업데이트                   
        
    def add_last(self, data: Any) -> None:
        
        if self.head is None:     # 연결리스트가 비어 있으면 add_first(data) 실행 
            self.add_first(data)  # !!!!!!!!!!!!!!!!!!!!클래스의 메서드 안에서 클래스 내부의 또 다른 메서드를 호출하는 법
            
        else:
            ptr = self.head              # 이것도 헤드노드를 참조하는 별도의 포인터 변수 ptr을 사용하는게 핵심
            while ptr.next is not None:  # 꼬리노드까지 간 다음 (i.e. ptr.next = None이 될때까지 간 다음)
                ptr = ptr.next
            ptr.next = self.current = Node(data, None)  # ptr은 꼬리노드를 참조, 따라서 기존 ptr.next는 None이었을 텐데, 
                                                        # ptr.next는 이제 None이 아닌 새로 추가되는 노드를 참조하게 만든다
            self.no += 1   # 전체 노드 개수 업데이트       
    
    def remove_first(self) -> None:
        
        if self.head is not None:  # 머리노드가 있다면 = 연결리스트가 비어있지 않다면 
            self.head = self.current = self.head.next  # self.head와 self.current의 참조를 self.head.next로 바꾼다
                                                       # i.e. 삭제전 머리노드는 어디에서도 참조되지 않는다
        self.no -= 1     # 전체 노드 개수 업데이트                                                                         
        
    def remove_last(self) -> None:
        
        if self.head is not None:   # 머리노드가 있다면 = 연결리스트가 비어있지 않다면 
            if self.head.next is None: # 연결리스트에 머리노드 하나만 존재하면
                self.remove_first()    # 머리노드 삭제
            else:
                ptr = self.head        # 헤드노드를 참조하는 별도의 포인터 변수 ptr과 함께
                pre = self.head        # 헤드노드를 참조하는 별도의 포인터 변수 pre도 사용하는게 핵심
                
                while ptr.next is not None: # 꼬리노드까지 간 다음 (i.e. ptr.next = None이 될때까지 간 다음)
                    pre = ptr               # pre는 ptr을 참조, while문을 다 돌고 나면 pre는 꼬리노드 직전노드를 참조
                    ptr = ptr.next          # ptr은 ptr.next를 참조 while문을 다 돌고 나면 ptr은 꼬리노드의 next, 즉 None을 참조 
                pre.next = None        # 꼬리노드의 직전노드의 next가 None을 참조, 즉 꼬리노드 삭제, 
                                    # 왜냐면 이렇게 되면 삭제전 꼬리노드는 어디에서도 참조되지 않는다. 즉 pre가 꼬리 노드가 된다.
                self.current = pre     
                self.no -= 1  # 전체 노드 개수 업데이트          
                
    def remove(self, p: Node) -> None:
        
        if self.head is not None:   # 머리노드가 있다면 = 연결리스트가 비어있지 않다면 
            if p is self.head:         # p가 참조하는 노드가 머리노드면
                self.remove_first()
            else:                      # p가 참조하는 노드가 머리노드가 아니면
                ptr = self.head        # 헤드노드를 참조하는 별도의 포인터 변수 ptr 사용
                
                while ptr.next is not p:  # p를 뒤쪽노드로 하는 노드(이 노드는 ptr이 참조)를 찾을때까지 반복
                    ptr = ptr.next
                    if ptr is None:    # 연결리스트 끝까지 가면
                                       # i.e. p가 참조하는 노드가 연결리스트에 존재하지 않으면 
                        return         # 함수 종료 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!! 
                ptr.next = p.next   # p 노드 삭제. 왜냐면 이렇게 되면 p가 참조하는 노드는 어디에서도 참조되지 않는다
                self.current = ptr  # self.current는 삭제된 p노드 앞에 있던 노드를 참조!
                self.no -= 1   # 전체 노드 개수 업데이트           
        
    def remove_current_node(self) -> None:
        
        self.remove(self.current) # !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
                                  # self.current 는 노드 자체가 아니라, 주목하는 노드를 참조하는 field 이다. 
    
    def clear(self) -> bool:
        while self.head is not None:  # 연결리스트 끝까지 가면서
            self.remove_first()       # 매번 머리노드 삭제
        self.current = None
        self.no = 0       
```

```python
class LinkedList:
    
    
    def __init__(self) -> None: # 노드가 하나도 없는 빈 연결리스트를 생성
        
        self.no = 0             # 연결리스트에 등록되어 있는 노드의 개수
        self.head = None        # self.head: 헤드 노드를 참조하기 위한 필드 
                                    # head는 머리 노드에 대한 참조일 뿐, 노드 그 자체가 아니다 !!!!!!!!!!!!!!!!!!
        self.current = None     # self.current : 현재 주목하고 있는 노드를 참조하는 필드 
                                # 주목 포인터라고 부르자.
                                    # 검색, 삽입, 삭제, 이동한 뒤의 해당 노드를 주목 (책 338p 표 8-1)
        
    def __len__(self) -> int:
        
        return self.no
    
    def search(self, data: Any) -> int:

        cnt = 0
        ptr = self.head              # 머리 노드부터 검색, 별도의 포인터 변수 ptr을 사용하는게 핵심.
        while ptr is not None:       # 연결리스트 끝까지 검색 (꼬리노드의 next는 None이므로)
            if ptr.data == data:     # 입력된 data 와 노드의 data가 일치하면
                self.current = ptr   # self.current를 해당 노드로 맞추고 
                return cnt           # 해당 노드의 순서번호를 리턴 (맨 앞 노드 순서번호는 0)
            cnt+=1                      # 일치하지 않으면 순서번호 +1
            ptr = ptr.next              # 머리노드의 뒤쪽노드를 참조
        
        return -1
    
    def __contains__(self, data: Any) -> Bool:  # 이 매직메서드를 구현함으로써 연결리스트에 in 연산자를 적용할 수 있다
        
        return self.search(data) >= 0
    
    def add_first(self, data: Any) -> None:
        
        ptr = self.head                            # 이것도 헤드노드를 참조하는 별도의 포인터 변수 ptr을 사용하는 게 핵심
        self.head = self.current = Node(data, ptr) # 입력받은 data 와 기존 헤드를 다음 노드로 하는 새 노드 생성 
                                                   # 새로 생성된 노드를 self.head와 self.current가 참조하게 한다.
        self.no += 1    # 전체 노드 개수 업데이트                   
        
    def add_last(self, data: Any) -> None:
        
        if self.head is None:     # 연결리스트가 비어 있으면 add_first(data) 실행 
            self.add_first(data)  # !!!!!!!!!!!!!!!!!!!!클래스의 메서드 안에서 클래스 내부의 또 다른 메서드를 호출하는 법
            
        else:
            ptr = self.head              # 이것도 헤드노드를 참조하는 별도의 포인터 변수 ptr을 사용하는게 핵심
            while ptr.next is not None:  # 꼬리노드까지 간 다음 (i.e. ptr.next = None이 될때까지 간 다음)
                ptr = ptr.next
            ptr.next = self.current = Node(data, None)  # ptr은 꼬리노드를 참조, 따라서 기존 ptr.next는 None이었을 텐데, 
                                                        # ptr.next는 이제 None이 아닌 새로 추가되는 노드를 참조하게 만든다
            self.no += 1   # 전체 노드 개수 업데이트       
    
    def remove_first(self) -> None:
        
        if self.head is not None:  # 머리노드가 있다면 = 연결리스트가 비어있지 않다면 
            self.head = self.current = self.head.next  # self.head와 self.current의 참조를 self.head.next로 바꾼다
                                                       # i.e. 삭제전 머리노드는 어디에서도 참조되지 않는다
        self.no -= 1     # 전체 노드 개수 업데이트                                                                         
        
    def remove_last(self) -> None:
        
        if self.head is not None:   # 머리노드가 있다면 = 연결리스트가 비어있지 않다면 
            if self.head.next is None: # 연결리스트에 머리노드 하나만 존재하면
                self.remove_first()    # 머리노드 삭제
            else:
                ptr = self.head        # 헤드노드를 참조하는 별도의 포인터 변수 ptr과 함께
                pre = self.head        # 헤드노드를 참조하는 별도의 포인터 변수 pre도 사용하는게 핵심
                
                while ptr.next is not None: # 꼬리노드까지 간 다음 (i.e. ptr.next = None이 될때까지 간 다음)
                    pre = ptr               # pre는 ptr을 참조, while문을 다 돌고 나면 pre는 꼬리노드 직전노드를 참조
                    ptr = ptr.next          # ptr은 ptr.next를 참조 while문을 다 돌고 나면 ptr은 꼬리노드의 next, 즉 None을 참조 
                pre.next = None        # 꼬리노드의 직전노드의 next가 None을 참조, 즉 꼬리노드 삭제, 
                                    # 왜냐면 이렇게 되면 삭제전 꼬리노드는 어디에서도 참조되지 않는다. 즉 pre가 꼬리 노드가 된다.
                self.current = pre     
                self.no -= 1  # 전체 노드 개수 업데이트          
                
    def remove(self, p: Node) -> None:
        
        if self.head is not None:   # 머리노드가 있다면 = 연결리스트가 비어있지 않다면 
            if p is self.head:         # p가 참조하는 노드가 머리노드면
                self.remove_first()
            else:                      # p가 참조하는 노드가 머리노드가 아니면
                ptr = self.head        # 헤드노드를 참조하는 별도의 포인터 변수 ptr 사용
                
                while ptr.next is not p:  # p를 뒤쪽노드로 하는 노드(이 노드는 ptr이 참조)를 찾을때까지 반복
                    ptr = ptr.next
                    if ptr is None:    # 연결리스트 끝까지 가면
                                       # i.e. p가 참조하는 노드가 연결리스트에 존재하지 않으면 
                        return         # 함수 종료 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!! 
                ptr.next = p.next   # p 노드 삭제. 왜냐면 이렇게 되면 p가 참조하는 노드는 어디에서도 참조되지 않는다
                self.current = ptr  # self.current는 삭제된 p노드 앞에 있던 노드를 참조!
                self.no -= 1   # 전체 노드 개수 업데이트           
        
    def remove_current_node(self) -> None:
        
        self.remove(self.current) # !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
                                  # self.current 는 노드 자체가 아니라, 주목하는 노드를 참조하는 field 이다. 
    
    def clear(self) -> bool:
        while self.head is not None:  # 연결리스트 끝까지 가면서
            self.remove_first()       # 매번 머리노드 삭제
        self.current = None
        self.no = 0       
        
    def next(self) -> bool:
        
        if self.current is None or self.current.next is None: # 연결리스트가 비었거나, 노드가 하나밖에 없으면
            False
        self.current = self.current.next
        return True
    
    def print_current_node(self) -> None:
        
        if self.current is None:
            print("노드가 존재하지 않습니다")
        else: print(self.current.data)
        
    def print(self) -> None:
        
        ptr = self.head
        
        while ptr is not None:  # 연결리스트 끝까지 검색 (꼬리노드의 next는 None이므로)
            print(ptr.data)
            ptr = ptr.next
```

# 문6) 이터러블, 이터레이터

**이터러블 객체는 원소를 1개씩 꺼내는 구조의 객체다.**

    어떤 객체가 이터러블이려면
    
    내장함수 iter()의 인수로 객체를 전달했을 때, 함수 iter()는 객체에 대한 이터레이터를 반환해야한다.
    
    따라서 이터러블 객체에는 내부에 __iter__ 매직 메서드가 구현되어야 한다.
        
**이터레이터는 데이터가 줄지어 늘어선 것을 표현하는 객체다.**

    이터레이터의 __next__() 함수를 호출하면, 또는 내장함수 next()에 이터레이터를 인수로 전달하면
    
    원소를 순차적으로 꺼내고, 다 꺼내면 StopIteration 예외처리를 보낸다
    
    따라서 이터레이터 객체에는 내부에 __iter__와 __next__ 매직 메서드가 구현되어야 한다

```python
from __future__ import annotations

class LinkedList:
    
    
    def __init__(self) -> None: # 노드가 하나도 없는 빈 연결리스트를 생성
        
        self.no = 0             # 연결리스트에 등록되어 있는 노드의 개수
        self.head = None        # self.head: 헤드 노드를 참조하기 위한 필드 
                                    # head는 머리 노드에 대한 참조일 뿐, 노드 그 자체가 아니다 !!!!!!!!!!!!!!!!!!
        self.current = None     # self.current : 현재 주목하고 있는 노드를 참조하는 필드 
                                # 주목 포인터라고 부르자.
                                    # 검색, 삽입, 삭제, 이동한 뒤의 해당 노드를 주목 (책 338p 표 8-1)
        
    def __len__(self) -> int:
        
        return self.no
    
    def search(self, data: Any) -> int:

        cnt = 0
        ptr = self.head              # 머리 노드부터 검색, 별도의 포인터 변수 ptr을 사용하는게 핵심.
        while ptr is not None:       # 연결리스트 끝까지 검색 (꼬리노드의 next는 None이므로)
            if ptr.data == data:     # 입력된 data 와 노드의 data가 일치하면
                self.current = ptr   # self.current를 해당 노드로 맞추고 
                return cnt           # 해당 노드의 순서번호를 리턴 (맨 앞 노드 순서번호는 0)
            cnt+=1                      # 일치하지 않으면 순서번호 +1
            ptr = ptr.next              # 머리노드의 뒤쪽노드를 참조
        
        return -1
    
    def __contains__(self, data: Any) -> Bool:  # 이 매직메서드를 구현함으로써 연결리스트에 in 연산자를 적용할 수 있다
        
        return self.search(data) >= 0
    
    def add_first(self, data: Any) -> None:
        
        ptr = self.head                            # 이것도 헤드노드를 참조하는 별도의 포인터 변수 ptr을 사용하는 게 핵심
        self.head = self.current = Node(data, ptr) # 입력받은 data 와 기존 헤드를 다음 노드로 하는 새 노드 생성 
                                                   # 새로 생성된 노드를 self.head와 self.current가 참조하게 한다.
        self.no += 1    # 전체 노드 개수 업데이트                   
        
    def add_last(self, data: Any) -> None:
        
        if self.head is None:     # 연결리스트가 비어 있으면 add_first(data) 실행 
            self.add_first(data)  # !!!!!!!!!!!!!!!!!!!!클래스의 메서드 안에서 클래스 내부의 또 다른 메서드를 호출하는 법
            
        else:
            ptr = self.head              # 이것도 헤드노드를 참조하는 별도의 포인터 변수 ptr을 사용하는게 핵심
            while ptr.next is not None:  # 꼬리노드까지 간 다음 (i.e. ptr.next = None이 될때까지 간 다음)
                ptr = ptr.next
            ptr.next = self.current = Node(data, None)  # ptr은 꼬리노드를 참조, 따라서 기존 ptr.next는 None이었을 텐데, 
                                                        # ptr.next는 이제 None이 아닌 새로 추가되는 노드를 참조하게 만든다
            self.no += 1   # 전체 노드 개수 업데이트       
    
    def remove_first(self) -> None:
        
        if self.head is not None:  # 머리노드가 있다면 = 연결리스트가 비어있지 않다면 
            self.head = self.current = self.head.next  # self.head와 self.current의 참조를 self.head.next로 바꾼다
                                                       # i.e. 삭제전 머리노드는 어디에서도 참조되지 않는다
        self.no -= 1     # 전체 노드 개수 업데이트                                                                         
        
    def remove_last(self) -> None:
        
        if self.head is not None:   # 머리노드가 있다면 = 연결리스트가 비어있지 않다면 
            if self.head.next is None: # 연결리스트에 머리노드 하나만 존재하면
                self.remove_first()    # 머리노드 삭제
            else:
                ptr = self.head        # 헤드노드를 참조하는 별도의 포인터 변수 ptr과 함께
                pre = self.head        # 헤드노드를 참조하는 별도의 포인터 변수 pre도 사용하는게 핵심
                
                while ptr.next is not None: # 꼬리노드까지 간 다음 (i.e. ptr.next = None이 될때까지 간 다음)
                    pre = ptr               # pre는 ptr을 참조, while문을 다 돌고 나면 pre는 꼬리노드 직전노드를 참조
                    ptr = ptr.next          # ptr은 ptr.next를 참조 while문을 다 돌고 나면 ptr은 꼬리노드의 next, 즉 None을 참조 
                pre.next = None        # 꼬리노드의 직전노드의 next가 None을 참조, 즉 꼬리노드 삭제, 
                                    # 왜냐면 이렇게 되면 삭제전 꼬리노드는 어디에서도 참조되지 않는다. 즉 pre가 꼬리 노드가 된다.
                self.current = pre     
                self.no -= 1  # 전체 노드 개수 업데이트          
                
    def remove(self, p: Node) -> None:
        
        if self.head is not None:   # 머리노드가 있다면 = 연결리스트가 비어있지 않다면 
            if p is self.head:         # p가 참조하는 노드가 머리노드면
                self.remove_first()
            else:                      # p가 참조하는 노드가 머리노드가 아니면
                ptr = self.head        # 헤드노드를 참조하는 별도의 포인터 변수 ptr 사용
                
                while ptr.next is not p:  # p를 뒤쪽노드로 하는 노드(이 노드는 ptr이 참조)를 찾을때까지 반복
                    ptr = ptr.next
                    if ptr is None:    # 연결리스트 끝까지 가면
                                       # i.e. p가 참조하는 노드가 연결리스트에 존재하지 않으면 
                        return         # 함수 종료 !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!! 
                ptr.next = p.next   # p 노드 삭제. 왜냐면 이렇게 되면 p가 참조하는 노드는 어디에서도 참조되지 않는다
                self.current = ptr  # self.current는 삭제된 p노드 앞에 있던 노드를 참조!
                self.no -= 1   # 전체 노드 개수 업데이트           
        
    def remove_current_node(self) -> None:
        
        self.remove(self.current) # !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
                                  # self.current 는 노드 자체가 아니라, 주목하는 노드를 참조하는 field 이다. 
    
    def clear(self) -> bool:
        while self.head is not None:  # 연결리스트 끝까지 가면서
            self.remove_first()       # 매번 머리노드 삭제
        self.current = None
        self.no = 0       
        
    def next(self) -> bool:
        
        if self.current is None or self.current.next is None: # 연결리스트가 비었거나, 노드가 하나밖에 없으면
            False
        self.current = self.current.next
        return True
    
    def print_current_node(self) -> None:
        
        if self.current is None:
            print("노드가 존재하지 않습니다")
        else: print(self.current.data)
        
    def print(self) -> None:
        
        ptr = self.head
        
        while ptr is not None:  # 연결리스트 끝까지 검색 (꼬리노드의 next는 None이므로)
            print(ptr.data)
            ptr = ptr.next


    def __iter__(self) -> LinkedListIterator: # 내장함수 iter()의 인수로 연결리스트 객체를 전달했을 때,
        
        return LinkedListIterator(self.head)   #  연결리스트 객체에 대한 이터레이터를 반환한다.
    
    
class LinkedListIterator:
    
    
    def __init__(self, head: Node):
        self.current = head
        
    def __iter__(self) -> LinkedListIterator:
        return self
    
    def __next__(self) -> Any:
        if self.current is None:
            raise StopIteration
        else:
            data = self.current.data
            self.current = self.current.next
            return data
```
