# 문3) KMP 법이란? 구현방법
패턴 안에 반복되는 부분이 있는 경우에 효과적인 문자열 검색법이다.  

검사대상인 텍스트에서 패턴을 비교해나가다가  
패턴의 반복되는 부분까지 일치하고   
그 나머지가 일치하지 않는 경우  
패턴에서 반복되는 부분을 건너 뛰고 비교를 이어나간다.  
  
예1)
    
    text    : ABCAE
    pattern : ABCABD
                  반복부분
    text : A  B  C |A| E
    pat  : A  B  C  A  B(불일치)
    text : A  B  C  A  E
    pat  :         _A_ B  C  A  B  D
           한칸 건너 뛰고 B부터 검사 재개
예2)

    text    : ABCABA
    pattern : ABCABD  
                   반복부분
    text : A  B  C |A  B| A  
     pat : A  B  C  A  B  D(불일치)
    text : A  B  C  A  B  A
     pat :         _A__B_ C  A  B  D
              두칸 건너 뛰고 C부터 검사 재개

    text    : AABD

    한 칸 건너뛰고 A부터 검사 재개

# 문4) 점프 테이블 만드는 법

    
    문자열 앞에
    패턴과 일치하는 일부 (패턴의 proper prefix)가
    뒤에서도 (패턴의 suffix로)  반복해서 나타날때

    패턴의 lps (longest prefix suffix)배열을 사용해 건너 뛸 수 있다

        A A B A A C
        0 1 0 1 2 0

        패턴의 lps 배열은
        방법1:proper prefix와 suffix가 일치하는지 여부로도 만들 수 있고
        방법2: 패턴을 오른쪽으로 시프트하면서 일치하는지 여부로도 만들 수 있다


    lps는 문자열중 suffix와 동일한 가장 긴 proper prefix입니다.
     여기서 proper prefix란 일반 prefix와 다르게 문자열 자기 자신은 제외됩니다.

        예) 문자열 "ABC"
            •	prefix : "", "A", "AB", "ABC"
            •	proper prefix: "", "A", "AB"

        문자열 "ACA"에서 lps를 구한다면 다음과 같습니다.
            •	proper prefix: "", "A", "AC"
            •	suffix: "", "A", "CA", "ACA"
        이중 suffix와 동일한 가장 긴 proper prefix인 lps는 "A" 입니다.


    lps 배열은 문자열 위치에 해당하는 하위 문자열의 lps 길이가 저장되어 있습니다.
    lps 배열을 사용하면 문자열에서 이미 검사했던 문자들을 다시 검사하는 것을 피할 수 있습니다.


    다음은 AABAAC 패턴에 대한 LPS배열의 단계별 구성 예입니다.


### step 1  
방법1  

    AABAAC
    A
    0
    하위 문자열 길이가 1인 경우 proper prefix는 없기 때문에 lps[0] 은 항상 0입니다.  
    (자기자신은 제외)  
방법2

    txt AABAAC
    pat AABAAC
        0
    패턴의 맨 처음과 텍스트의 첫문자가 일치하면 한칸 밀고, 
    불일치해도 한칸 밀어야 하므로 
    점프할 필요가 없다. 
    점프는 반복되는 부분을 만났을 때 하는 것이다.
    
### step2
방법1

    AABAAC
    AA
    01
    하위 문자열 "AA"의 lps는 "A"이기 때문에 lps[1]=1을 저장합니다.
    proper prefix : A
    suffix : A, AA

방법2

    A|A|BAAC
     |A|ABAAC
    0 1
    반복되는 부분이 문자 하나이므로 한칸 점프해 비교를 재개한다.

### step3
방법1

    AABAAC 
    AAB
    010
    하위 문자열 "AAB"의 lps는 없기 때문에 lps[2] = 0을 저장합니다.
    proper prefix : A, AA
    suffix : AB, B AAB
방법2

    AABAAC
      AABAAC
    010
    반복되는 부분이 없으므로 점프 없이 오른쪽으로 한칸 밀어 처음부터 비교한다.
    
### step 4
방법1

    AABAAC
    AABA
    0101
    하위 문자열 "AABA"의 lps는 "A"이기 때문에 lps[3] = 1을 저장합니다.
    proper prefix : A, AA, AAB
    suffix : A, BA, ABA, AABA
방법2

    AAB|A|AC
       |A|ABAAC
    010 1
    반복되는 부분의 문자가 하나이므로 한칸 건너뛰어 패턴의 두번째 A부터 검사를 재개한다.
    

### step 5
방법1

    AABAAC
    AABAA
    01012
    하위 문자열 "AABAA"의 lps는 “A”, “AA"이기 때문에 lps[4] = 2를 저장합니다.
    proper prefix : A, AA, AAB, AABA
    suffix : A, AA, BAA, ABAA, AABAA
방법2

    AAB|AA|C
       |AA|BAAC
    010 12
    반복되는 부분이  AA이므로 두칸 건너 뛰어서 B부터 비교를 재개한다.
### step 6
방법1

    AABAAC
    AABAAC
    010120
    하위 문자열 "AABAAC"의 lps는 없기 때문에 lps[5]=0을 저장합니다.
    proper prefix : A, AA, AAB, AABA, AABAA
    suffix : C, AC, AAC, BAAC, ABAAC, AABAAC
    
방법2

    AABAAC
         AABAAC
    010120
    반복되는 부분이 없으므로 점프 없이 오른쪽으로 한칸 밀어 처음부터 비교한다.
    

# 문5) KMP 구현방법

```python
def kmp_match(txt: str, pat: str) -> int:
    # 패턴에서 점프 테이블 생성
    pt = 1  # 패턴의 두번째 문자부터 비교 시작 (첫번째 문자에서는 점프할 필요가 없으므로)
    pp = 0  # 패턴의 첫번째 문자로 비교 시작
    
    skip = [0] * (len(pat) + 1) # 점프 테이블의 길이가 패턴 길이+1 인 이유
                                # 검색코드에서 텍스트커서와 패턴커서의 각 인덱스가 0에서 시작하고                        
                                # 텍스트와 패턴의 일치여부를 따진 다음에 커서 인덱스를 +1해주므로
                                # 점프테이블에서 실제 점프할 위치를 기록하는 시작위치는 점프테이블의 두번쨰 인덱스부터여야 한다
    skip[pt] = 0
    while pt != len(pat):
        if pat[pt] == pat[pp]: # (패턴의 두번째 문자에서부터 패턴의 첫번째 문자와 비교를 시작해) 일치하는 경우
            pt += 1
            pp += 1
            skip[pt] = pp  # 점핑되는 칸 수를 기록
        elif pp == 0:
            pt += 1
            skip[pt] = pp  # 일치하지 않고 pp가 0인 경우 다음 위치로 이동
        else:
            pp = skip[pp]  # 일치하지 않는 경우 점프 테이블에 따라 pp를 조정

    # 텍스트에서 패턴 검색
    pt = pp = 0 # 텍스트커서, 패턴커서 0으로 초기화
    while pt != len(txt) and pp != len(pat):
        if txt[pt] == pat[pp]:
            pt += 1
            pp += 1  # 문자가 일치하는 경우 계속해서 다음 문자를 비교
        elif pp == 0:  # 패턴과 일치하는 부분이 처음부터 계속 없으면 pt만 밀고 나감
            pt += 1  
        else:
            pp = skip[pp]  # 패턴과 일치하는 부분이 있었다가 불일치하면, 점프 테이블에 따라 pp를 조정

    return pt - pp if pp == len(pat) else -1  # 패턴이 전부 일치했다면 시작 인덱스 반환, 그렇지 않으면 -1 반환
```
